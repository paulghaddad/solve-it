```python
# write a function that takes

# (1) a directory name 
# (2) a function

# it returns a dict whose keys are the filenames in the 
#   directory, and whose values are the result of running
#   a function on each argument

# we can ignore errors/exceptions
```

```python
from os.path import join
import glob

def apply_func_to_dir(dirname, func):
    output = {}
    
    for one_filename in glob.glob(join(dirname, '*')):
        try:
            output[one_filename] = func(one_filename)
        except:
            pass
    
    return output

apply_func_to_dir('/etc/', len)
```

```python
def get_file_length(filename):
    total = 0
    for one_line in open(filename):
        total += len(one_line)
    return total
```

```python
apply_func_to_dir('/etc/', get_file_length)
```
