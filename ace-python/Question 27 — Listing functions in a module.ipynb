```python
for one_attrname in dir(random):
    attr = getattr(random, one_attrname)
    print(f'{one_attrname:20} {str(attr):30}: {type(attr).__name__}')
```

```python
[one_attrname
for one_attrname in dir(random)
if type(getattr(random, one_attrname)) == type(random.randint)]
```
