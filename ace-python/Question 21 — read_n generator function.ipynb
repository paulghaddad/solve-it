```python
# Write a generator function, read_n, that takes two arguments --
# a filename and a number (n).  In each iteration, the 
# generator should return n lines from the file.  If there aren't
# n lines remaining, then it should return as many lines as possible.

def read_n(filename, n):
    f = open(filename)
    
    while True:
        one_chunk = ''.join(f.readline()
                            for i in range(n))
        if one_chunk:  # if we didn't get an empty string
            yield one_chunk
        else:
            break


for one_chunk in read_n('/etc/passwd', 3):
    print(one_chunk)
```
